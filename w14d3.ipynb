Importiamo il modulo math e proviamo a usare le funzioni
.sin() .cos() .factorial()
e la variabile
.pi
la riconoscete?

```python
import math
angolo = 90

pigreco = math.pi
radianti = angolo * pigreco / 180 #converto in radianti
print(radianti)
seno = math.sin(radianti)
coseno = math.cos(math.floor(radianti))
fattoriale = math.factorial(math.floor(radianti))

print(f'Seno: {seno}  Coseno: {coseno} Fattoriale: {fattoriale} ')
```

L'azienda Object SpA ha creato una lista di quanti oggetti ha venduto ogni mese nell'ultimo anno:
lst = [2000, 5500, 7200, 4320, 1280, 1900, 2500, 3900, 6410, 8150, 7100, 5350] trasformiamola in un array NumPy (casting):
lst = np.array(lst) e rispondiamo alle domande del CEO della Object SpA:
• qual è stata la vendita massima mensile? E quella minima?
• quali sono le vendite mensili maggiori di 4999 oggetti? E quante ne sono?
• quali sono le vendite minori di 3000 oggetti?
• in media quanti oggetti sono stati venduti al mese?

```python
import numpy as np
import pandas as pd
lst = [2000, 5500, 7200, 4320, 1280, 1900, 2500, 3900, 6410, 8150, 7100, 5350]
lst = np.array(lst)


#troviamo la vendita massima
vendita_max = lst.max()
print(f'Vendita massima: {vendita_max}')

#troviamo la vendita minima
vendita_min = lst.min()
print(f'Vendita minima: {vendita_min}')

#troviamo le vendite superiori a 4999
vendite_maggiori_4999 = lst[lst > 4999]
print(f'Vendite maggiori di 4999: {vendite_maggiori_4999}')

#contiamo il numero di vendite superiori a 4999 oggetti
numero_vendite_maggiori_4999 = len(vendite_maggiori_4999)
print(f'Numero di vendite maggiori di 4999 oggetti:{numero_vendite_maggiori_4999}')

#quali sono le vendite minori di 3000 oggetti?
vendite_minori_3000 = lst[lst < 3000]
print(f'Vendite minori di 3000 oggetti: {vendite_minori_3000}')

#media oggetti venduti mensilmente
media_oggetti_venduti = lst.mean()
print(f'Media degli oggetti venduti mensilmente: {media_oggetti_venduti}')

#aggiungo un data frame per rappresentare i dti
dati = [list(["Vendita Massima", vendita_max]),
        list(["Vendita Minima", vendita_min]),
        list(["Vendite maggiori di 4999", vendite_maggiori_4999]),
        list(["Numero di vendite maggiori di 4999 oggetti", numero_vendite_maggiori_4999]),
        list(["Vendite minori di 3000 oggetti", vendite_minori_3000]),
        list(["Media degli oggetti venduti mensilmente", media_oggetti_venduti])]
df = pd.DataFrame(dati)
df.columns = ('Attributo', 'Valore')
df
```
